### a. Dataset Cleaning & Selection

```python
import pandas as pd
```

```python
# Load the dataset
file_path = "../data/apartments_for_rent_classified_100K.csv"  
df = pd.read_csv(file_path, delimiter=";", encoding="ISO-8859-1", low_memory=False)  

display(df)
```

```python
df['time'] = pd.to_datetime(df['time'], unit='s')
display(df.head())
```

```python
df.to_csv("../data/raw.csv", index=False)
```

```python
# Remove features
column_removed = df.drop(['category', 'address'], axis=1)
display(column_removed.head())
```

```python
# Change the empty value in amenities to No amenities
column_removed['amenities'] = column_removed['amenities'].fillna('No amenties')
column_removed['amenities'] = column_removed['amenities'].apply(lambda x: x if isinstance(x, str) and x.strip() != '' else 'No amenities')

# Change the empty value in pets_allowed to No pets_allowed
column_removed['pets_allowed'] = column_removed['pets_allowed'].replace('Cats,Dogs,None', 'Cats,Dogs')
column_removed['pets_allowed'] = column_removed['pets_allowed'].fillna('No pets allowed')

display(column_removed.head())
```

```python
# Drop all the na value now
df_cleaned = column_removed.dropna()
display(df_cleaned)
```

```python
duplicate_ids = df_cleaned[df_cleaned.duplicated(subset=['id'], keep='first')]
print(f"Number of duplicate IDs: {len(duplicate_ids)}")
print("Duplicate IDs:")
print(duplicate_ids)
```

```python
df_cleaned = df_cleaned.drop_duplicates(subset='id', keep='first')
```

```python
duplicate_ids = df_cleaned[df_cleaned.duplicated(subset=['id'], keep='first')]
print(f"Number of duplicate IDs: {len(duplicate_ids)}")
```

```python
# Checked, it's cleaned
df_cleaned.to_csv("../data/dataset.csv", index=False)
```

### c. Data Subset: Divide to different tables for databases

```python
# Use the new dataset
apartments = pd.read_csv('../data/final_dataset.csv')
```

```python
all_amenities = apartments['amenities'].str.split(',').explode().unique()
all_amenities
```

```python
# amenities
amenities_list = sorted(list(all_amenities))  
amenities_df = pd.DataFrame({
    'amenity_id': range(1, len(amenities_list) + 1),
    'amenity_name': amenities_list
})
display(amenities_df)
amenities_df.to_csv('../data/amenities.csv', index=False)
```

```python
# property_amenities
amenity_map = {name: aid for name, aid in zip(amenities_list, range(1, len(amenities_list) + 1))}
records = []
for _, row in apartments.iterrows():
    apt_id = row['id']
    amenities_field = row['amenities']
    for a in amenities_field.split(','):
        name = a.strip()
        if name in amenity_map:
            records.append({'id': apt_id, 'amenity_id': amenity_map[name]})

amenities_types_df = pd.DataFrame(records)
display(amenities_types_df)
amenities_types_df.to_csv('../data/amenities_types.csv', index=False)
```

```python
# pricing
pricing = apartments[['id', 'price', 'currency']]
display(pricing.head())
pricing.to_csv("../data/pricing.csv", index=False)
```

```python
# price_details
payment = apartments[['id', 'price_display', 'price_type']]
display(payment.head())
payment.to_csv("../data/payment.csv", index=False)
```

```python
# pets
sources = apartments[['id', 'pets_allowed', 'fee']]
display(sources.head())
sources.to_csv("../data/pets.csv", index=False)
```
